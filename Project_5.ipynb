### Data Cleaning & Preparation

```python
import pandas as pd

# Load and clean the data
df = pd.read_csv("C:/Users/ACER/Downloads/order_dataset.csv")
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df.drop_duplicates(inplace=True)

# Add return flag
df['Is Return'] = df['Final Quantity'] < 0
```

### Overall Return Rate

```python
return_rate = df['Is Return'].mean() * 100
print(f"Overall Return Rate: {return_rate:.2f}%")
```

### Return Rate by Category

```python
category_return = df.groupby('Category')['Is Return'].mean().sort_values(ascending=False) * 100
print(category_return)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(x=category_return.index, y=category_return.values)
plt.xticks(rotation=60)
plt.ylabel('Return Rate (%)')
plt.title('Return Rate by Category')
plt.tight_layout()
plt.show()
```

### Top Returned Products

```python
top_returns = df[df['Is Return']].groupby(['Item Code', 'Item Name']).size().sort_values(ascending=False).head(10).reset_index(name = 'Return Count')
print(top_returns.to_markdown(index=False))
```

### Top Returning Customers

```python
top_customers = df[df['Is Return']].groupby('Buyer ID').size().sort_values(ascending=False).head(10).reset_index(name = 'Return Count')
print(top_customers.to_markdown(index=False))
```

### Monthly Return Trend

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_returns = df.groupby('Month')['Is Return'].mean() * 100

monthly_returns.plot(kind='line', figsize=(10, 5), marker='o', title='Monthly Return Rate (%)')
plt.ylabel('Return Rate (%)')
plt.grid(True)
plt.show()
```

### Total Refunds by Category

```python
refunds_by_category = df.groupby('Category')['Refunds'].sum().sort_values(ascending=False)

refunds_by_category.plot(kind='bar', figsize=(10,5), title='Total Refunds by Category')
plt.ylabel('Total Refunds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Return Rate by Discount Band

```python
def discount_band(x):
    if x <= -50:
        return 'High Discount'
    elif -49 <= x <= -10:
        return 'Moderate Discount'
    else:
        return 'Low/No Discount'

df['Discount Band'] = df['Price Reductions'].apply(discount_band)
discount_group = df.groupby('Discount Band')['Is Return'].mean() * 100

sns.barplot(x=discount_group.index, y=discount_group.values)
plt.ylabel('Return Rate (%)')
plt.title('Return Rate by Discount Band')
plt.show()
```

###  High-Risk Products Export

```python
product_stats = (
    df.groupby(['Item Code', 'Item Name'])
    .agg(total_orders=('Item Code', 'count'),
         returns=('Is Return', 'sum'),
         return_rate=('Is Return', 'mean'),
         total_refunded=('Refunds', 'sum'))
    .reset_index()
)

high_risk = product_stats[(product_stats['total_orders'] > 5) & (product_stats['returns'] > 1)]
high_risk.loc[:, 'return_rate'] = high_risk['return_rate'] * 100
high_risk.sort_values(by='return_rate', ascending=False).to_csv('high_risk_products.csv', index=False)
```
